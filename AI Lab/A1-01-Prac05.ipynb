### Minimax Algo
function minimax(node, depth, isMaximizingPlayer, alpha, beta):

if node is a leaf node :

return value of the node

if isMaximizingPlayer :

bestVal = -INFINITY

for each child node :

value = minimax(node, depth+1, false, alpha, beta)

bestVal = max( bestVal, value)

alpha = max( alpha, bestVal)

if beta <= alpha:

break

return bestVal

else :

bestVal = +INFINITY

for each child node :

value = minimax(node, depth+1, true, alpha, beta)

bestVal = min( bestVal, value)

beta = min( beta, bestVal)

if beta <= alpha:

break

return bestVal

```python
MIN =-9999999
MAX = 9999999
def minimax(depth, index, maximizingPlayer,values, alpha, beta):
    if depth == 3:#if node is a leaf node then return value of the node
        return values[index]
    if maximizingPlayer:
        bestVal = -9999999
        # for each child node 
        for i in range(0, 2):
            val = minimax(depth + 1, index * 2 + i,False, values, alpha, beta)
            bestVal = max(bestVal, val)
            alpha = max(alpha, bestVal)
            if beta <= alpha:
                 break
        return bestVal

    else:
        bestVal = 9999999
        #for each child node 
        for i in range(0, 2):
            val = minimax(depth + 1, index * 2 + i,True, values, alpha, beta)
            bestVal = min( bestVal, val)
            beta = min( beta, bestVal)
            if beta <= alpha:
                break
        return bestVal


if __name__ == "__main__":
    values = [3,5,6,9,1,2,0,-1]
    print("The value is :", minimax(0, 0, True, values, MIN, MAX))
```

```python
import random
# Human vs intelligent computer tic tac toe
def print_tic_tac_toe(values):
    print("\n")
    print("\t     |     |")
    print("\t  {}  |  {}  |  {}".format(values[0], values[1], values[2]))
    print('\t_____|_____|_____')
 
    print("\t     |     |")
    print("\t  {}  |  {}  |  {}".format(values[3], values[4], values[5]))
    print('\t_____|_____|_____')
 
    print("\t     |     |")
 
    print("\t  {}  |  {}  |  {}".format(values[6], values[7], values[8]))
    print("\t     |     |")
    print("\n")

def isMovesLeft(values):
    for i in range(9):
            if (values[i] == ' '):
                return True
    return False
 
def check_win(player_pos, cp):
 
    soln = [[1, 2, 3], [4, 5, 6], [7, 8, 9], [1, 4, 7], [2, 5, 8], [3, 6, 9], [1, 5, 9], [3, 5, 7]]
 
    for x in soln:
        if all(y in player_pos[cp] for y in x):
            return True
    return False       
 

def check_draw(player_pos):
    if len(player_pos['X']) + len(player_pos['O']) == 9:
        return True
    return False       
   
def evaluate(player_pos, opponent):
    soln = [[1, 2, 3], [4, 5, 6], [7, 8, 9], [1, 4, 7], [2, 5, 8], [3, 6, 9], [1, 5, 9], [3, 5, 7]]
 
    for x in soln:
        if all(y in player_pos[opponent] for y in x):
            if opponent =='X':
                return -10
            if opponent == 'O':
                return 10
    return 0 

def alpha_beta_pruning(player_pos, cp, values, depth, isMax, alpha, beta):
    if cp == 'X':
        opponent ='O'
    else:
        opponent ='X'
        
    score = evaluate(player_pos, opponent)
   
    if score == 10:
        return score - depth

    if score == -10:
        return score + depth

    if not isMovesLeft(values):
        return 0

    if isMax:
        best = -float('inf')

        for i in range(9):
                if values[i] == ' ':
                    values[i] = 'O'
                    best = max(best, alpha_beta_pruning(player_pos, cp, values, depth + 1, not isMax, alpha, beta))
                    values[i] = ' '
                    alpha = max(alpha, best)

                    if beta <= alpha:
                        break
          
        return best
    else:
        best = float('inf')

        for i in range(9):
                if values[i] == ' ':
                    values[i] = 'O'
                    best = min(best, alpha_beta_pruning(player_pos, cp, values, depth + 1, not isMax, alpha, beta))
                    values[i] = ' '
                    beta = min(beta, best)

                    if beta <= alpha:
                        break
      
        return best
    
def findBestMove(player_pos, cp, values):
    bestVal = -float('inf')
    bestMoves = []
    
    for i in range(9):
            if values[i] == ' ':
                values[i] = 'X'
                moveVal = alpha_beta_pruning(player_pos, cp, values, 0, False, -float('inf'), float('inf'))
                values[i] = ' '

                if moveVal > bestVal:
                    bestVal = moveVal
                    bestMoves = [i]
                elif moveVal == bestVal:
                    bestMoves.append(i)

    return random.choice(bestMoves)
  

def validate_move(move,values):
    if move < 1 or move > 9:
            print("Wrong Input!!! Try Again")
            return False
    if values[move-1] != ' ':
            print("Place already filled. Try again!!")
            return False
    else:        
        return True
    

def single_game(cp,cur_player, player1, player2):
    values = [' ' for x in range(9)]
    player_pos = {'X':[], 'O':[]}
    while True:
        print_tic_tac_toe(values)
         
        if cur_player == player1:
                print("Player ", cp, " turn. Which box? : ", end="")
                move = int(input()) 
                if validate_move(move,values) == False:
                     continue
        if cur_player == player2:
                print("Computer's turn. Which box? : ", end="")
                move = findBestMove(player_pos, cp, values)
                if validate_move(move,values) == False:
                     continue
 
        values[move-1] = cp
 
        player_pos[cp].append(move)
 
        if check_win(player_pos, cp):
            print_tic_tac_toe(values)
            print("Player ", cp, " has won the game!!")     
            print("\n")
            return cur_player
 
        if check_draw(player_pos):
            print_tic_tac_toe(values)
            print("Game Drawn")
            print("\n")
            return 'D'
        if cur_player == player1:
            cur_player = player2
        else:
            cur_player = player1
        if cp == 'X':
            cp = 'O'
        else:
            cp = 'X'
            
            
if  __name__ == "__main__":
    print("Player 1")
    player1 = input("Enter the name : ")
    print("\n")
 
    print("Player 2 : Computer")
    player2 = "Computer"
    print("\n")
     
    cur_player = player1
    player_choice = {'X' : "", 'O' : ""}
 
    options = ['X', 'O']
 
 
    while True:
 
        print("Turn to choose for", cur_player)
        print("Enter 1 for X #player to always choose X")
        print("Enter 2 for O#Computer always O")
        print("Enter 3 to Quit")
 
        try:
            choice = int(input())   
        except ValueError:
            print("Wrong Input!!! Try Again\n")
            continue
  
        if choice == 1:
            player_choice['X'] = cur_player
            if cur_player == player1:
                player_choice['O'] = player2
            else:
                player_choice['O'] = player1
 
        elif choice == 2:
            player_choice['O'] = cur_player
            if cur_player == player1:
                player_choice['X'] = player2
            else:
                player_choice['X'] = player1
         
        elif choice == 3:
            break  
         
        else:
            print("Wrong Choice!!!! Try Again\n")
        
        winner = single_game(options[choice-1],cur_player, player1, player2)
         # Switch player who chooses X or O
        if cur_player == player1:
            cur_player = player2
        else:
            cur_player = player1
        
```
